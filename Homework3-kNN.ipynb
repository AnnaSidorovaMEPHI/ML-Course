# kNN and wine-dataset

### В этом задании вам нужно подобрать оптимальное значение k для алгоритма kNN. Будем использовать набор данных Wine, где требуется предсказать сорт винограда, из которого изготовлено вино, используя результаты химических анализов.

```python
import numpy as np
import pandas as pd
```

```python
columns=['Wine','Alcohol','Malic acid','Ash','Alcalinity of ash','Magnesium','Total phenols','Flavanoids','Nonflavanoid phenols','Proanthocyanins','Color intensity','Hue','OD280/OD315 of diluted wines','Proline']
```

```python
df=pd.read_csv(r'./wine.data',sep=',', names=columns)
```

```python
df.info()
```

```python
df.head()
```

```python
X=df[columns[1:]]
y=df.Wine #Выделение целевой переменной
```

### Вычислить качество на всех разбиениях можно при помощи функции sklearn.model_selection.cross_val_score. Результатом является массив, значения которого нужно усреднить.

```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
m=0
for k in range(1,51):
    kNN = KNeighborsClassifier(n_neighbors=k)
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    j=(cross_val_score(kNN, X, y, scoring='accuracy',cv=kf)).mean() #Усредняем, чтобы оценить качество
    if j>m:
        kmax=k
        m=j
    print("k =",k,":",cross_val_score(kNN, X, y, scoring='accuracy', cv=kf)) 
```

### Найдите точность классификации на кросс-валидации для метода k ближайших соседей (sklearn.neighbors.KNeighborsClassifier), при k от 1 до 50. При каком k получилось оптимальное качество? Чему оно равно (число в интервале от 0 до 1)? Данные результаты и будут ответами на вопросы 1 и 2.


```python
print(kmax,m) # результат
```
